```
import requests
from bs4 import BeautifulSoup
import re
```

```

# query_text = 'mobile phone'
# query_text = query_text.replace(' ', '+')
# number = 60

def get_img_urls(query_text, n_pics):
    
    number = 0
    query_text = query_text.replace(' ', '+')

    total_img_links = []
    while number <= n_pics:
        MAIN_URL = f'https://th.images.search.yahoo.com/search/images?norw=1&ei=UTF-8&fr=sfp&fr2=p%3As%2Cv%3Ai&o=js&p={query_text}&tmpl=&nost=1&b={number}&iid=Y.6&ig=0af185d75d2e4fb88600000000d23bd5&rand=1613516875071'
        # MAIN_URL = f'https://th.images.search.yahoo.com/search/images?ei=UTF-8&fr=sfp&fr2=sb-top-th.images.search&o=js&p={query_text}&tmpl=&nost=1&b={number}&iid=Y.6&ig=0af0253438b54cd79000000000867a90&rand=1613506083391'
        r = requests.get(MAIN_URL)
        soup = BeautifulSoup(r.content,'lxml')
        results = soup.find_all('img',{'src':re.compile('.*')})
        img_links = [x['src'].replace("\\", "") for x in results]
        total_img_links = total_img_links + img_links
        number += 60
    return total_img_links
```

```
import pandas as pd
import numpy as np
import requests
import os
from tqdm.notebook import tqdm
import concurrent.futures

import random
import string




def get_random_string(length):
    # choose from all lowercase letter
    letters = string.ascii_lowercase
    result_str = ''.join(random.choice(letters) for i in range(length))
    return result_str


original_dir = '/content'
os.chdir(original_dir)
def download_img(img_link):
    r = requests.get(img_link)
    file_name = get_random_string(20) + '.jpg'
    # file_path = os.path.join(img_dest_path,file_name) 
    with open(file_name, 'wb') as f:
        # for chunk in r.iter_content(chunk_size = 128):
        #     f.write(chunk)
        f.write(r.content)

def scrape_and_download_imgs(search_queries, n_pics = 100):
    original_dir = os.getcwd()

    main_folder = 'scraped_image'
    os.makedirs(main_folder, exist_ok = True)
    for query in tqdm(search_queries):
        os.chdir(original_dir)
        current_folder = os.path.join(main_folder, query)
        os.makedirs(current_folder, exist_ok = True)
        os.chdir(current_folder)

        img_link = get_img_urls(query,n_pics = n_pics)
        
        with concurrent.futures.ThreadPoolExecutor() as executor:
            executor.map(download_img, img_link)
    
    os.chdir(original_dir)
```
